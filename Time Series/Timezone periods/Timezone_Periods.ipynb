```python
import pandas as pd
df = pd.read_csv('wmt.csv')
df
```

```python
df.set_index('Line Item', inplace = True)
df
```

```python
df = df.T 
df
```

```python
df.index = pd.PeriodIndex(df.index, freq = 'Q-JAN')
df.index
```

```python
df['Start Date'] = df.index.map(lambda x: x.start_time)
df
```

```python
df['End Date'] = df['End Date'].dt.date
df
```
